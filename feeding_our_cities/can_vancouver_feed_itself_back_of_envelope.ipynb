```python
# back-of-envelope
#- https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=3210035801


import matplotlib.pyplot as plt

KG_PER_HUNDREDWEIGHT = 45.36
ACRES_PER_SQKM = 1000000 / 4046

KCAL_PER_PERSON_DAY = 2000
DAYS_PER_YEAR = 365

average_yield_potatoes_hundredweight_per_harvested_acres = 330 # 300 hundredweight / acre

average_yield_potatoes_tonnes_per_hectare = 35.66

population_of_greater_vancouver = 3000000
population_of_vancouver = 662248
population_of_guelph = 150000
population_of_jabodetabek = 35 * 1000 * 1000
population_of_jabodetabek = 11 * 1000 * 1000

area_of_greater_vancouver_sqkm = 2878.93
area_of_vancouver_sqkm = 115.18
area_of_guelph_sqkm = 87.22
area_of_jabodetabek_sqkm = 7000
area_of_jabodetabek_sqkm = 661

# more conservative estimate, only counting usable area
area_of_greater_vancouver_sqkm = 1920.25

average_yield_potatoes_kg_per_sqkm = average_yield_potatoes_hundredweight_per_harvested_acres * KG_PER_HUNDREDWEIGHT * ACRES_PER_SQKM
average_yield_potatoes_kg_per_sqkm = average_yield_potatoes_tonnes_per_hectare * 1000 * 100
```

```python
potato_kcal_per_100g = 87
potato_kcal_per_kg = 87 * 1000 / 100

kcal_potatoes_per_sqkm = average_yield_potatoes_kg_per_sqkm * potato_kcal_per_kg

# more conservative estimate
kcal_potatoes_per_sqkm = 9.2 * 1000000 * ACRES_PER_SQKM

total_kcal_potential_greater_vancouver = kcal_potatoes_per_sqkm * area_of_greater_vancouver_sqkm
```

```python
kcal_requirement_of_greater_vancouver = population_of_greater_vancouver * KCAL_PER_PERSON_DAY * DAYS_PER_YEAR
```

```python
plt.bar(x=["KCAL potential", "KCAL requirement"], height=[total_kcal_potential_greater_vancouver, kcal_requirement_of_greater_vancouver])
```

```python
total_kcal_potential_vancouver = kcal_potatoes_per_sqkm * area_of_vancouver_sqkm
kcal_requirement_of_vancouver = population_of_vancouver * KCAL_PER_PERSON_DAY * DAYS_PER_YEAR

plt.bar(x=["KCAL potential", "KCAL requirement"], height=[total_kcal_potential_vancouver, kcal_requirement_of_vancouver])
```
