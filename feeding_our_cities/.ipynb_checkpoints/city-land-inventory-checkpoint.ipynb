```python
import geopandas
from geodatasets import get_path

import matplotlib.pyplot as plt

# pull the same data with several different sources
# csv
# kml
# shapefile
# geojson
```

```python
gdf = geopandas.read_file('geofiles/Street_Centrelines.geojson')
```

```python
columns_to_keep = ["STREET", "STREETNAME", "STATUS", "PLACE", "SEQUENCE", "SEGTYPE", "ROADWIDTH", "ROADLENGTH", "Shape__Length", "ROADCLASS", "OPSCLASS", "OWNEDBY", "TRUCKROUTE", "geometry"]


important_city_roads_gdf = gdf.loc[
    (gdf["PLACE"] == "Guelph") 
    & (gdf["OWNEDBY"].isin(["City of Guelph", "University of Guelph"]))
    & (gdf["ROADCLASS"].isin(["Arterial", "Collector", "Expressway"])), columns_to_keep]

all_city_roads_gdf = gdf.loc[
    (gdf["PLACE"] == "Guelph") 
    & (gdf["OWNEDBY"].isin(["City of Guelph", "University of Guelph"])), columns_to_keep]
```

```python
tree_gdf = geopandas.read_file('geofiles/Trees.geojson')
```

```python
gdf3 = geopandas.read_file('geofiles/Parks.geojson')
```

```python
city_parks_gdf = gdf3.loc[gdf3["Ownership"] == 'CITY OF GUELPH', :]
```

```python
gdf4 = geopandas.read_file('geofiles/Buildings.geojson')
```

```python
buildings_area = gdf4.Shape__Area.sum()
```

```python
city_area = 87430000.0
```

```python
gdf5 = geopandas.read_file('geofiles/Property_Lines.geojson')
```

```python
# land available for common use farming

public_road_surface_parcels_guelph_gdf = gdf5.loc[
    (gdf5["MUNICIPALI"] == "CITY OF GUELPH")
    & (gdf5["PARCELSTAT"] == "ACTIVE")
    & (gdf5["OWNERSHIP"].isin(["CITY OF GUELPH", "GUELPH HYDRO", "UNIVERSITY OF GUELPH", "EDUCATION BOARD WELLINGTON", "EDUCATION BOARD GUELPH"]))
    & (gdf5["PARCELTYPE"].isin(["ROAD PARCEL", "SURFACE PARCEL"])), :]

public_road_parcels_guelph_gdf = gdf5.loc[
    (gdf5["MUNICIPALI"] == "CITY OF GUELPH")
    & (gdf5["PARCELSTAT"] == "ACTIVE")
    & (gdf5["OWNERSHIP"].isin(["CITY OF GUELPH", "GUELPH HYDRO", "UNIVERSITY OF GUELPH", "EDUCATION BOARD WELLINGTON", "EDUCATION BOARD GUELPH"]))
    & (gdf5["PARCELTYPE"].isin(["ROAD PARCEL"])), :]
```

```python
# get all the private land

private_parcels_guelph_gdf = gdf5.loc[
    (gdf5["MUNICIPALI"] == "CITY OF GUELPH")
    & (gdf5["PARCELSTAT"] == "ACTIVE")
    & (gdf5["OWNERSHIP"].isin(["PRIVATE"]))
    & (gdf5["PARCELTYPE"].isin(["ROAD PARCEL", "SURFACE PARCEL"])), :]
```

```python
all_parcels_in_guelph = gdf5.loc[(gdf5["MUNICIPALI"] == "CITY OF GUELPH") & (gdf5["PARCELSTAT"] == "ACTIVE"), :]
```

```python
buildings_in_guelph_gdf = gdf4.loc[gdf4["MUN"] == 'GUELPH', :]
```

```python
area_of_buildings_in_guelph = float(buildings_in_guelph_gdf["Shape__Area"].sum()) * 2

tree_area = float(len(tree_gdf)) * 200
tree_area = 0

area_of_public_land = float(public_road_surface_parcels_guelph_gdf["Shape__Area"].sum())
area_of_private_land = float(private_parcels_guelph_gdf["Shape__Area"].sum())
area_of_important_city_roads = float((important_city_roads_gdf["ROADWIDTH"] * important_city_roads_gdf["ROADLENGTH"]).sum())
area_of_all_city_roads = float((all_city_roads_gdf["ROADWIDTH"] * all_city_roads_gdf["ROADLENGTH"]).sum())

area_of_guelph = 87430000.0
area_of_guelph_2 = float(all_parcels_in_guelph["Shape__Area"].sum())

area_of_all_public_road_parcels = float(public_road_parcels_guelph_gdf["Shape__Area"].sum())
park_areas_guelph = float(city_parks_gdf["Shape__Area"].sum())
```

```python
print("TOTAL AREA")
print("#############")
print("Area of Guelph(1) \t\t: ", round(area_of_guelph))
print("Area of Guelph(2) \t\t: ", round(area_of_guelph_2))
print("Area of private land \t\t: ", round(area_of_private_land))
print("Area of public land \t\t: ", round(area_of_public_land))

print("UNUSABLE AREA")
print("#############")
print("Area of important city roads \t\t: ", round(area_of_important_city_roads))
print("Area of all public road parcels \t: ", round(area_of_all_public_road_parcels))
print("Area of buildings in Guelph \t\t: ", round(area_of_buildings_in_guelph))
print("Area of trees in Guelph \t\t: ", round(tree_area))
print("Area of public parks \t\t\t: ", round(park_areas_guelph))

free_area = (area_of_public_land + area_of_private_land - area_of_important_city_roads - area_of_buildings_in_guelph - park_areas_guelph - tree_area)

print("USABLE AREA")
print("##############")
print("Open unstructured area in Guelph \t\t: ", round(free_area))

print("PERCENTAGE USABLE AREA")
print("######################")
print("\nLand use ratio \t\t: ", round(free_area / area_of_guelph_2, 2))
```
