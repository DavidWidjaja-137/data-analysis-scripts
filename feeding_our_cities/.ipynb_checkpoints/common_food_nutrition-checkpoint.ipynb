# (temporary) Read the CNF database

Create several useful functions to read the CNF database

```python
import pandas as pd
import numpy as np
```

```python
# read the food name table

food_name_useful_cols = ["FoodID", "FoodCode", "FoodGroupID", "FoodDescription"]
food_name_useful_cols = ["FoodID", "FoodGroupID", "FoodDescription"]


food_name_df = pd.read_csv('./cnf_2015/FOOD NAME.csv', encoding_errors='replace', usecols=food_name_useful_cols)
```

```python
# read the food group table

food_group_table_cols = ["FoodGroupID", "FoodGroupCode", "FoodGroupName"]
#food_group_table_cols = ["FoodGroupID", "FoodGroupName"] 

food_group_table_df = pd.read_csv('./cnf_2015/FOOD GROUP.csv', encoding_errors='replace', usecols=food_group_table_cols, index_col='FoodGroupID')

# Dairy and Egg Products
# Spices and Herbs
# Fats and Oils
# 
basic_and_vegetarian_food_groups = [1, 2, 4, 9, 11, 12, 16, 20]

food_group_table_df = food_group_table_df[food_group_table_df["FoodGroupCode"].isin(basic_and_vegetarian_food_groups)]
```

```python
# read and combine the yield amount and yield name tables

yield_amount_cols = ["FoodID", "YieldID", "YieldAmount"]

yield_amount_df = pd.read_csv('./cnf_2015/YIELD AMOUNT.csv', encoding_errors='replace', usecols=yield_amount_cols)

yield_name_cols = ["YieldID", "YieldDescription"]
dtypes = {"YieldID": np.int64}

yield_name_df = pd.read_csv('./cnf_2015/YIELD NAME.csv', nrows=79, encoding_errors='replace', usecols=yield_name_cols, dtype=dtypes, index_col='YieldID')

yield_df = yield_amount_df.join(yield_name_df, on="YieldID")
```

```python
# read and combine the conversion factor tables

conversion_factor_cols = ["FoodID", "MeasureID", "ConversionFactorValue"]

conversion_factor_df = pd.read_csv(
    './cnf_2015/CONVERSION FACTOR.csv', encoding_errors='replace', usecols=conversion_factor_cols,
)

measure_name_cols = ["MeasureID", "MeasureDescription"]

measure_name_df = pd.read_csv('./cnf_2015/MEASURE NAME.csv', encoding_errors='replace', usecols=measure_name_cols, index_col="MeasureID")

conversion_df = conversion_factor_df.join(measure_name_df, on="MeasureID")
```

```python
# read and combine the refuse tables

refuse_amount_cols = ["FoodID", "RefuseID", "RefuseAmount"]

refuse_amount_df = pd.read_csv('./cnf_2015/REFUSE AMOUNT.csv', encoding_errors='replace', usecols=refuse_amount_cols)

refuse_name_cols = ["RefuseID", "RefuseDescription"]

refuse_name_df = pd.read_csv('./cnf_2015/REFUSE NAME.csv', encoding_errors='replace', nrows=150, usecols=refuse_name_cols, index_col="RefuseID")

refuse_df = refuse_amount_df.join(refuse_name_df, on="RefuseID")
```

```python
# read the nutrient amount and nutrient name tables

#nutrient_name_cols = ["NutrientID", "NutrientCode", "NutrientSymbol", "NutrientUnit"]
nutrient_name_cols = ["NutrientID", "NutrientCode", "NutrientName", "NutrientUnit"]

nutrient_name_df = pd.read_csv('./cnf_2015/NUTRIENT NAME.csv', usecols=nutrient_name_cols, encoding_errors='replace')

nutrient_name_df = nutrient_name_df[nutrient_name_df['NutrientID'].isin([203, 208])]

nutrient_amount_cols = ["FoodID", "NutrientID", "NutrientValue"]

nutrient_amount_df = pd.read_csv('./cnf_2015/NUTRIENT AMOUNT.csv', encoding_errors='replace', usecols=nutrient_amount_cols)

nutrient_df = nutrient_amount_df.merge(nutrient_name_df, on="NutrientID")

nutrient_calorie_df = nutrient_df.loc[nutrient_df["NutrientID"] == 208, :]
nutrient_protein_df = nutrient_df.loc[nutrient_df["NutrientID"] == 203, :]
```

```python
test_calorie_df = (
    food_name_df.merge(food_group_table_df, on="FoodGroupID", how='inner')
    .merge(refuse_df, on="FoodID", how='left')
    .merge(conversion_df, on="FoodID", how='left')
    .merge(yield_df, on="FoodID", how='left')
    .merge(nutrient_calorie_df, on="FoodID", how='left')    
    .drop(columns=["FoodGroupID", "RefuseID", "MeasureID", "YieldID", "NutrientID", "NutrientCode"])
    .rename(columns={'NutrientValue': 'NutrientValue100g'})
)

test_calorie_df["NutrientValuePortion"] = test_calorie_df["NutrientValue100g"] * test_calorie_df["ConversionFactorValue"]

test_calorie_df.to_csv('food_calories.csv', index=False)
```

```python
test_protein_df = (
    food_name_df.merge(food_group_table_df, on="FoodGroupID", how='inner')
    .merge(refuse_df, on="FoodID", how='left')
    .merge(conversion_df, on="FoodID", how='left')
    .merge(yield_df, on="FoodID", how='left')
    .merge(nutrient_protein_df, on="FoodID", how='left')    
    .drop(columns=["FoodGroupID", "RefuseID", "MeasureID", "YieldID", "NutrientID", "NutrientCode"])
    .rename(columns={'NutrientValue': 'NutrientValue100g'})
)

test_protein_df["NutrientValuePortion"] = test_protein_df["NutrientValue100g"] * test_protein_df["ConversionFactorValue"]

test_protein_df.to_csv('food_protein.csv', index=False)
```
